## A-Preprocess for NH

```python
StateGroups = MergedCounty.groupby('State')
```

```python
ReportNH = ['Illinois', 'Texas', 'Virginia', 'California', 'New Jersey',  'Oregon']
# NoHisRaceNum_States = []

L = []


for state, State in StateGroups:
    if state in ReportNH:
        
        State.loc[State.index, 'CvdVax_ZWhiteNum'] = State.loc[State.index, 'Numb_White']
        
        
        State.loc[State.index, 'CvdVax_WhiteNum'] = np.nan
        State.loc[State.index, 'CvdVax_NHWhiteNum'] = State.loc[State.index, 'Numb_White']
        
        State.loc[State.index, 'CvdVax_BlackNum'] = State.loc[State.index, 'Numb_Black']
        
        State.loc[State.index, 'ReportNH'] = 1
        
    else:
        # State = StateGroups.get_group('California')
        State.loc[State.index, 'CvdVax_ZWhiteNum'] = State.loc[State.index, 'Numb_White']
        
        State.loc[State.index, 'CvdVax_WhiteNum'] = State.loc[State.index, 'Numb_White']
        State.loc[State.index, 'CvdVax_NHWhiteNum'] = np.nan
        
        State.loc[State.index, 'CvdVax_BlackNum'] = State.loc[State.index, 'Numb_Black']
        
        State.loc[State.index, 'ReportNH'] = 0
    
    # print(State.columns)
    State = State.drop(['Numb_White', 'Numb_Black'], axis = 1)
    
    
    try:
        if state in ReportNH:
            State.loc[State.index, 'FullCvdVax_ZWhiteNum'] = State.loc[State.index, 'FullNumb_White']
        
            State.loc[State.index, 'FullCvdVax_WhiteNum'] = np.nan
            State.loc[State.index, 'FullCvdVax_NHWhiteNum'] = State.loc[State.index, 'FullNumb_White']
            State.loc[State.index, 'FullCvdVax_BlackNum'] = State.loc[State.index, 'FullNumb_Black']

        else:
            State.loc[State.index, 'FullCvdVax_ZWhiteNum'] = State.loc[State.index, 'FullNumb_White']
        
            State.loc[State.index, 'FullCvdVax_WhiteNum'] = State.loc[State.index, 'FullNumb_White']
            State.loc[State.index, 'FullCvdVax_NHWhiteNum'] = np.nan
            State.loc[State.index, 'FullCvdVax_BlackNum'] = State.loc[State.index, 'FullNumb_Black']
            
        # print(State.columns)
        State = State.drop(['FullNumb_White', 'FullNumb_Black'], axis = 1)
        
    except:
        print('No Full Data')

    L.append(State)
    
StateCvdVaxData = pd.concat(L)
```

```python
# All

# Z-Type
StateCvdVaxData['CvdVax_ZWhiteRate']   = StateCvdVaxData['CvdVax_ZWhiteNum'] / StateCvdVaxData['NHWhiteNum'] * 100

# Report-Type and NH-Type
StateCvdVaxData['CvdVax_WhiteRate']   = StateCvdVaxData['CvdVax_WhiteNum']   / StateCvdVaxData['WhiteNum'] * 100
StateCvdVaxData['CvdVax_NHWhiteRate'] = StateCvdVaxData['CvdVax_NHWhiteNum'] / StateCvdVaxData['NHWhiteNum'] * 100

# M-Type
target_col = 'CvdVax_MWhiteRate'
StateCvdVaxData[target_col] = np.nan
for ethn in ['', 'NH']:
    col = 'CvdVax_' + ethn + 'WhiteRate'
    # print(col)
    index = - StateCvdVaxData[col].isna()
    StateCvdVaxData.loc[index, target_col] = StateCvdVaxData.loc[index, col] 

# Try to be Close to Black
StateCvdVaxData['CvdVax_MBlackRate']   = StateCvdVaxData['CvdVax_BlackNum']   / StateCvdVaxData['BlackNum'] * 100


try:
    # All
    
    # Z-Type
    StateCvdVaxData['FullCvdVax_ZWhiteRate']   = StateCvdVaxData['FullCvdVax_ZWhiteNum']   / StateCvdVaxData['NHWhiteNum'] * 100

    StateCvdVaxData['FullCvdVax_WhiteRate']   = StateCvdVaxData['FullCvdVax_WhiteNum']   / StateCvdVaxData['WhiteNum'] * 100
    StateCvdVaxData['FullCvdVax_NHWhiteRate'] = StateCvdVaxData['FullCvdVax_NHWhiteNum'] / StateCvdVaxData['NHWhiteNum'] * 100

    # Try to be Close to NHWhite
    target_col = 'FullCvdVax_MWhiteRate'
    StateCvdVaxData[target_col] = np.nan
    for ethn in ['', 'NH']:
        col = 'FullCvdVax_' + ethn + 'WhiteRate'
        # print(col)
        index = - StateCvdVaxData[col].isna()
        StateCvdVaxData.loc[index, target_col] = StateCvdVaxData.loc[index, col] 

    # Try to be Close to Black
    StateCvdVaxData['FullCvdVax_MBlackRate']   = StateCvdVaxData['FullCvdVax_BlackNum']   / StateCvdVaxData['BlackNum'] * 100
    
except:
    print('No Full Data')
```

## C-Filter Valid

```python
def valid_preprocess(StateCvdVaxData):
    target_list = [i for i in StateCvdVaxData.columns if ('MWhite' in i or 'MBlack' in i) and 'Full' not in i]
    print(target_list)
    # Filter 1
    critieria_nonan = {}
    critieria_nonan_list = target_list
    for col in target_list:
        # print(col)
        critieria_nonan[col] = (-StateCvdVaxData[col].isna()).astype(int)


    # Filter 2
    cirteria_range = {}
    cirteria_range_list = [i for i in StateCvdVaxData.columns if 'MWhite' in i or 'MBlack' in i]
    
    for col in target_list:
        cirteria_range[col] = (StateCvdVaxData[col] <= 100).astype(int)

    StateCvdVaxData['Valid'] = 1
    for critieria in [critieria_nonan, cirteria_range]:
        for col, critierion in critieria.items():
            StateCvdVaxData['Valid'] = StateCvdVaxData['Valid']*critierion
    
    return StateCvdVaxData 

StateCvdVaxData = valid_preprocess(StateCvdVaxData)

StateCvdVaxData = StateCvdVaxData[StateCvdVaxData['Valid'] == 1]
```
